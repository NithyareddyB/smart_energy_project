# Module 2: Exploratory Data Analysis (EDA)
This notebook explores the raw dataset from NASA POWER, Open-Meteo, and the grid demand CSV.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load data
df_solar = pd.read_csv('../data/raw/nasa_solar_raw.csv', index_col=0, parse_dates=True)
df_weather = pd.read_csv('../data/raw/weather_raw.csv', index_col=0, parse_dates=True)
df_demand = pd.read_csv('../data/raw/demand_raw.csv', index_col=0, parse_dates=True)
```

```python
# Time plots
plt.figure(figsize=(14,4))
plt.plot(df_demand['demand_mw'].tail(24*14))
plt.title('Hourly Grid Demand (last 14 days)')
plt.ylabel('MW')
plt.show()
```

```python
# Correlation heatmap
merged = df_solar.join(df_weather, how='inner').join(df_demand, how='inner')
plt.figure(figsize=(10,8))
sns.heatmap(merged.corr(), cmap='coolwarm', annot=False)
plt.title('Feature Correlation')
plt.show()
```

```python
# Missing value analysis
print('Missing values per column:')
print(merged.isnull().sum())
```
