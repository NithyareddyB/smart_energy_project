```python
from src.preprocessing.cleaner import DataCleaner
from src.preprocessing.feature_eng import FeatureEngineer
from src.preprocessing.scaler import DataScaler
from src.preprocessing.splitter import DataSplitter
```

```python
cleaner = DataCleaner()
df_clean = cleaner.clean()
df_clean.head()
```

```python
engineer = FeatureEngineer()
df_features = engineer.transform(df_clean)
df_features.head()
```

```python
scaler = DataScaler()
df_scaled = scaler.fit_transform(df_features)
df_scaled.describe()
```

```python
splitter = DataSplitter()
splits = splitter.split(df_scaled)
print('Train X shape:', splits['X_train'].shape)
print('Val X shape:', splits['X_val'].shape)
print('Test X shape:', splits['X_test'].shape)
```
